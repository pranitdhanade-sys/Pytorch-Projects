```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt 
```

```python
torch.manual_seed(42)
num_smaples = 100
x0 = torch.randn(num_smaples, 2) + 1.5
x1 = torch.randn(num_smaples, 2) - 1.5
x = torch.cat((x0, x1), dim=0)
y = torch.cat((torch.ones(num_smaples,1),torch.zeros(num_smaples,1)), dim =0)
```

```python
class LogisticRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear() = nn.linear(2,1)
    
    def forward(self, x):
        return torch.sigmoid(self.linear(x))
model = LogisticRegression()
```

```python
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
epochs = 200
for epoch in range(epochs):
    y_pred = model(x)
    loss = criterion(y_pred,y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1)%20 == 0:
        print(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.4f}")
```

```python
plt.figure(fisize=(7,5))
plt.scatter(x0[:,0],x0[:,1], c='blue', label='Class 1')
plt.scatter(x1[:,0],x1[:,1], c='red', label='Class 0')
```

```python
xx, yy = torch.meshgrid(torch.linspace(-4,4,100), torch.linspace(-4,4,100))
grid = torch.cat((xx.reshape(-1,1),yy.reshape(-1,1)), dim=1)
preds = model(grid)
zz = preds.reshape(100,100).detach()
```

```python
plt.countourf(xx,yy,zz,levels=[0,0.5,1],alpha=0.3,colors=['red','blue'])
plt.legend()
plt.title("Logistic Regression Decision Boundary")
plt.show()
```
