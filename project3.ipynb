```python
import torch 
import torch.nn
import torch.optim as optim
import matplotlib.pyplot as plt 
```

```python
x = torch.linspace(0,10,50).unsqueeze(1)
y = 2*x+3+torch.randn(x.size()) * 0.5
```

```python
class LinearRegressionModel(nn.Model):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1,1)

    def forward(self, x):
        return self.linear(x)

model = LinearRegressionModel
```

```python
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 200
for epoch un range(epochs):
    y_pred = model(x)

    loss = criterion(y_pred,y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1)%20 == 0:
    [w,b] = model.parameters()
    print(f"Epoch {epoch+1}/{epochs} | Loss: {loss.item():.4f} | W: {w.item():.4f}, b: {b.item():.4f}")
```

```python
[w,b] = model.parameters()
print("\n Final learned parameters:")
print(f"W = {w.item():.3f}, b={b.item():.3f}")
```

```python
test_x = torch.tensor([[12.0]])
pred_y = model(test_x)
print(f"Prediction for x = 12 y={pred_y.item:.3f}")
```

```python
plt.scatter(x.detach(), y.detach(), label="Data")
plt.plot(x.detach(),model(x).detach(),color='red',label="Fitted Line")
plt.legend()
plt.title("Linear Regression using torch.nn")
plt.show()
```
