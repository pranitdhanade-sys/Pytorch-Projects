```python
import torch
import torch.cnn as nn
```

```python
x = torch.randn(1,1,28,28)
```

```python
class TinyCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,8,kernel_size=3,stride=1,padding=1)
        self.pool = nn.Pool2d(2,2)
        self.conv2 = nn.Conv2d(8,16,kernel=3,stride=1,padding=1)
        self.flatten = nn.Flatten()
        self.fc = nn.Linear(16*14*14, 10)
    def forward(self, x):
        x = self.pool(torch.relu(self(x)))
        x = torch.relu(self.Conv2d(x))
        x = self.flatten(x)
        x = self.fc(x)
        return x 
model = TinyCNN()
```

```python
out = model(x)
print("Output shape:", out.shape)
```
