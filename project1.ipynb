```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True)
print("x=",x)

y = 3* x**2 + 2*x +1
print("y=",y)
```

```python
y.backward()
print("Gradient dy/dx", x.grad)
```

```python
x = torch.tensor([1.0,2.0,3.0], require_grad=True)
y = x**2+2*x
out = y.sum()
out.backward()
```
